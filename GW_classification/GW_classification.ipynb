```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('ml_pasta.csv')
df
```

Thresholds were selected in accordance with those previously used and discussed in several CONUS wide studies of unburned watersheds (Johnson et al., 2020, Hare et al., 2021), and more targeted watershed scale studies that leveraged observations from multiple ecoregions (Briggs et al., 2018). 

Specifically, sites with   amplitude ratios greater than 0.65 and phase lags < 10 days were considered to have an air-coupled signal, sites with amplitude ratios < 0.65 and phase lags < 10 days were considered to have a deep groundwater signal, and any sites with phase lags greater than 10 days were considered to have a shallow groundwater signal


```python

df.plot(y='amp', x='ph_l', kind='scatter', color='dodgerblue') 

plt.axvline(x=10, color='r', linestyle='--' )  # Vertical line at x=10
#plt.axhline(y=0.65, color='r', linestyle='--' )  # Horizontal line at y=0.65

# Define the x-range for the horizontal line
min_value = df['ph_l'].min()-4
x_start = min_value 
x_end = 10
x = np.linspace(x_start, x_end, 100)  # Create an array of x-values

# Define the y-value for the horizontal line
y_value = 0.65
y = np.full_like(x, y_value)  # Create an array of y-values with the same length as x, filled with y_value


plt.plot(x, y, color='red', linestyle='--', label=f'y = {y_value}') 

plt.xlim(min_value, df['ph_l'].max()+4)

plt.ylabel('Amplitude ratio')
plt.xlabel('Phase lag (days)')
plt.title(' ')


plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np  # For potential color mapping

# Sample DataFrame (replace with your actual data)
data = pd.DataFrame({
    'x': [1, 2, 3, 4, 5],
    'y': [5, 3, 6, 2, 7],
    'color_col': ['A', 'B', 'A', 'C', 'B'],  # Column to color by (categorical)
    'value_col': [10, 20, 15, 25, 18]       # Column to color by (numerical)
})

data = df

# 1. Coloring by a Categorical Column:

# Get unique categories and assign colors
unique_categories = data['labels'].unique()
colors = plt.cm.get_cmap('viridis', len(unique_categories))  # Choose a colormap

# Create a mapping from category to color
color_map = {category: colors(i) for i, category in enumerate(unique_categories)}

# Scatter plot with colors based on the 'color_col'
plt.figure(figsize=(8, 6))
for category in unique_categories:
    subset = data[data['labels'] == category]
    plt.scatter(subset['ph_l'], subset['amp'], label=category, color=color_map[category])

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title(' ')
plt.legend()
plt.grid(True)
plt.show()
```
