```python
# Import necessary libraries
import matplotlib.pyplot as plt
import os
import pandas as pd
from google.colab import drive
print("Mounting Google Drive...") # 1. Mount Google Drive
drive.mount('/content/drive/') # This step will prompt you to authorize Google Colab to access your Google Drive.
print("Google Drive mounted successfully!")
from datetime import datetime
# Get the current time
current_time = datetime.now()
# Print the current time
print("Current time:", current_time)
import glob
```

Define the path to your CSV files

```python
csv_folder_path = '/content/drive/My Drive/LSTM_DOY/LSTM_results_DOY_grid_search'
```

```python

# output best model results
best_KGE_result_csv='/content/drive/My Drive/LSTM_DOY/LSTM_results_DOY_grid_search/best_KGE/best_KGE_result_' +   current_time.strftime("%Y%m%d%H%M") + ".csv"
```

```python







# List all CSV files in the directory
csv_files = glob.glob(os.path.join( csv_folder_path, "*.csv"))

# List to hold individual DataFrames
df_list = []

# Loop through CSV files and read into DataFrames
for file in csv_files:
    df = pd.read_csv(file)
    df_list.append(df)

# Concatenate all DataFrames into a single DataFrame
combined_df = pd.concat(df_list, ignore_index=True)

# Optionally, display the resulting DataFrame
print(combined_df.head())
```

```python


# Count of each unique item in the 'fruit' column
counts = combined_df['model_site'].value_counts()

print(counts)
```

```python
result = combined_df.loc[combined_df.groupby('model_site')['KGE'].idxmax()]


print(result.sort_values(by=[  'KGE' , 'model_site' ]))

result.to_csv(best_KGE_result_csv)
```
